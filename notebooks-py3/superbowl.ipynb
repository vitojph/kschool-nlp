```python
tweets = []
RUTA = '../data/2017-superbowl-tweets.tsv'
for line in open(RUTA).readlines():
    tweets.append(line.split('\t'))
```

```python
from textblob import TextBlob

for tweet in tweets:
    try:
        t = TextBlob(tweet[3]) # in Python2: t = TextBlob(tweet[3].decode('utf-8'))
        if t.sentiment.polarity < -0.5:
            print(tweet[3], '-->', t.sentiment)
    except IndexError:
        pass
```

```python
for tweet in tweets:
    try:
        t = TextBlob(tweet[3]) # in Python2: t = TextBlob(tweet[3].decode('utf-8'))
        print(" ".join(t.noun_phrases))
    except IndexError:
        pass
```
